# Data Preprocessing (Data Cleaning, Feature Engineering)
- Menggunakan 1 teknik data preprocessing dan menjelaskan alasan pemilihan teknik yang digunakan dengan tepat (3 poin) #
- Menggunakan 1 teknik data preprocessing tambahan dan menjelaskan alasan pemilihan teknik yang digunakan dengan tepat (2 poin) #

```python
# Penggabungan Atribut

df['total'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['average'] = df['total']/3
df['result'] = df['average'].apply(lambda x: x > 60)
display(df)

# Lakukan dengan pandas
# Hal ini dilakukan supaya mempermudah menentukan status kelulusan siswa
```

nilai matematika, membaca, dan menulis digabung ke "total", hitung rata-rata sebagai "average", tentukan kelulusan jika rata-rata > 60 dan simpan sebagai "result". Jika memiliki nilai diatas 60, maka anak tersebut dianggap lulus.

```python
# Teknik preprocessing pertama dengan feature selection

df = df.drop(columns=['lunch', 'math_score', 'reading_score', 'writing_score', 'total', 'average'])
display(df)
```

Menghapus kolom yang tidak dibutuhkan yaitu luch, math score, reading score, writing score, total dan average

```python
# One Hot Encoding

A = df.iloc[:, 0].values
B = df.iloc[:, 1].values
C = df.iloc[:, 2].values
D = df.iloc[:, 3].values
E = df.iloc[:, 4].values

print(df)

data = {
    "gender": pd.Series(A, index=range(1000)),
    "race_ethnicity": pd.Series(B, index=range(1000)),
    "parental_level_of_education": pd.Series(C, index=range(1000)),
    "test_preparation_course": pd.Series(D, index=range(1000)),
    "result": pd.Series(E, index=range(1000)),
}

df = pd.DataFrame(data)

# Lakukan one-hot encoding pada 4 kolom pertama
df = pd.get_dummies(df, columns=['gender', 'race_ethnicity', 'parental_level_of_education', 'test_preparation_course'])

display(df)
```

Karena fitur diatas tidak dalam bentuk numerik maka dilakukan one hot encoding mengubah data kategorikal menjadi bentuk numerik

# Pelatihan, penyetelan model (model tuning)
- Kesesuaian pemilihan algoritma machine learning dengan masalah yang dihadapi (4 poin)
- Membagi data menjadi data untuk pelatihan dan data untuk diuji (2 poin)
- Menggunakan K-fold cross-validation (2 poin)

- Melakukan pencarian hyperparameter terbaik untuk algoritma yang digunakan (2 poin)
- Menggunakan Grid Search untuk mencari kombinasi hyperparameter terbaik dari sebuah model (5 poin)

### 1. split data

```python
df_copy = df.copy()
df_copy.drop('result', axis=1, inplace=True)

X = df_copy
y = df['result']


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)


display(X_train)
display(y_train)
display(X_test)
display(y_test)
```

### 2. menggunakan stratefied k-fold dan gridsearch

Menggunakan stratefied k-fold dikarenakan dapat menjaga distribusi kelas yang seimbang dalam setiap lipatan (fold) yang dibentuk selama proses validasi silang. Sedangkan Gridsearch disini digunakan untuk menemukan hyperparameter terbaik dengan akurasi tertinggi

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

### 3. Logistic Regression

Cocok digunakan untuk klasifikasi nilai selain itu, algoritma ini tidak membutuhkan waktu banyak untuk dilatih

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

# Define the parameter grid for grid search
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],  # Regularization strength
    'penalty': ['l1', 'l2'],       # Regularization type
    'solver': ['liblinear'],       # Solver for L1 regularization (useful for sparse data)
}
grid_search = GridSearchCV(lr, param_grid, cv=skf, scoring='accuracy')

grid_search.fit(X_train, y_train)

print("Best parameters:", grid_search.best_params_)
```

```python
best_lr = grid_search.best_estimator_
y_pred_lr = best_lr.predict(X_test)
display(y_pred_lr)
```

### 4. AdaBoost
Salah satu model yang cocok juga untuk klasifikasi juga cocok untuk dataset di mana proporsi kelulusan dan ketidaklulusan tidak seimbang.

```python
from sklearn.ensemble import AdaBoostClassifier

# Define the AdaBoost classifier
ada = AdaBoostClassifier()

# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 150, 200],
    'learning_rate': [0.01, 0.1, 0.5, 1]
}

grid_search = GridSearchCV(estimator=ada, param_grid=param_grid, cv=5, scoring='accuracy')

grid_search.fit(X_train, y_train)
print("Best parameters:", grid_search.best_params_)
```

```python
best_ada = grid_search.best_estimator_

y_pred_ada = best_ada.predict(X_test)
display(y_pred_ada)
```

### 5. Gradient Boosting
Model yang cocok digunakan untuk klasifikasi dan regresi dan efektif pada data dengan hubungan yang kompleks (data yang besar dengan jumlah fitur yang banyak) dan non-linier.

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier(random_state=42)

param_grid = {
    'n_estimators': [100, 200, 300],  # Number of trees
    'learning_rate': [0.01, 0.1, 0.2],  # Learning rate
    'max_depth': [3, 4, 5],  # Maximum depth of trees
    'min_samples_split': [2, 4, 6],  # Minimum samples to split a node
    'min_samples_leaf': [1, 2, 4],  # Minimum samples in a leaf
    'subsample': [0.8, 1.0],  # Fraction of samples to use for each tree
}

grid_search = GridSearchCV(gb, param_grid, cv=skf, scoring='accuracy')

grid_search.fit(X_train, y_train)
print("Best parameters:", grid_search.best_params_)
```

```python
best_gb = grid_search.best_estimator_

y_pred_gb = best_gb.predict(X_test)
display(y_pred_gb)
```

# Pengukuran Kinerja atau Evaluasi Model
- Melakukan pengukuran kinerja (performance measure) dengan metode yang tepat dan disertai penjelasan yang tepat (10 poin)

Menggunakan Cross Val Score

```python
from sklearn.model_selection import cross_val_score
models = [best_lr, best_ada, best_gb]

def crossval(models):
    data = {
    'Models': ['Logistic regression','AdaBoost', 'Gradient Boosting'],
    'Accuracy': [],
    'Precision': [],
    'Recall': [],
    'F1': []}
    for model in models:
        accuracy = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy').mean()
        precision = cross_val_score(model, X_train, y_train, cv=5, scoring='precision').mean()
        recall = cross_val_score(model, X_train, y_train, cv=5, scoring='recall').mean()
        f1 = cross_val_score(model, X_train, y_train, cv=5, scoring='f1').mean()

        data['Accuracy'].append(accuracy)
        data['Precision'].append(precision)
        data['Recall'].append(recall)
        data['F1'].append(f1)
    return data


data = crossval(models)
df = pd.DataFrame(data)

display(df)
```

### Kesimpulan

Dalam Train, Logistic regression memiliki akurasi yang paling tinggi, Gradient boosting memiliki presisi yang paling tinggi dan adaboost memiliki recall dan f1 paling tinggi.

```python
import pandas as pd
from sklearn.metrics import log_loss, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Define y_test and y_pred_lr, y_pred_ada, y_pred_gb

# Calculate metrics
log_loss_lr = log_loss(y_test, y_pred_lr)
log_loss_ada = log_loss(y_test, y_pred_ada)
log_loss_gb = log_loss(y_test, y_pred_gb)

accuracy_lr = accuracy_score(y_test, y_pred_lr)
accuracy_ada = accuracy_score(y_test, y_pred_ada)
accuracy_gb = accuracy_score(y_test, y_pred_gb)

precision_lr = precision_score(y_test, y_pred_lr)
precision_ada = precision_score(y_test, y_pred_ada)
precision_gb = precision_score(y_test, y_pred_gb)

recall_lr = recall_score(y_test, y_pred_lr)
recall_ada = recall_score(y_test, y_pred_ada)
recall_gb = recall_score(y_test, y_pred_gb)

f1_lr = f1_score(y_test, y_pred_lr)
f1_ada = f1_score(y_test, y_pred_ada)
f1_gb = f1_score(y_test, y_pred_gb)

# Create a DataFrame
metrics_df = pd.DataFrame({
    'Model': ['Logistic Regression', 'AdaBoost', 'Gradient Boosting'],
    'Log Loss': [log_loss_lr, log_loss_ada, log_loss_gb],
    'Accuracy': [accuracy_lr, accuracy_ada, accuracy_gb],
    'Precision': [precision_lr, precision_ada, precision_gb],
    'Recall': [recall_lr, recall_ada, recall_gb],
    'F1-Score': [f1_lr, f1_ada, f1_gb],
})

# Print the DataFrame
display(metrics_df)
```
